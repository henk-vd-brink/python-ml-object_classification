```python
import random
import shutil
from PIL import Image
from matplotlib import pyplot as plt
import numpy as np
import torchvision.transforms as transforms
import torch.nn as nn
```

```python
def get_labels(file_uri):
    with open(file_uri, "r") as annotation_file:
        annotations = annotation_file.readlines()
        
    split_annotations = [annotation.split(" ") for annotation in annotations]
    return [(split_annotation[0], split_annotation[1]) for split_annotation in split_annotations]
```

```python
train_part = 0.8
val_part = 0.1
test_part = 0.1

labels = get_labels("../data/processed/raw/_annotations.txt")

train_labels = []; val_labels = []; test_labels = []


for i, label in enumerate(labels):
    if (random_value:=random.uniform(0,1)) <= train_part:
        train_labels.append(label)
    elif train_part < random_value <= train_part+val_part:
        val_labels.append(label)
    else:
        test_labels.append(label)
        
print("Number of training labels:", len(train_labels)/len(labels))
print("Number of validation labels:",len(val_labels)/len(labels))
print("Number of testing labels:",len(test_labels)/len(labels))
```

```python
for label_type in ["train", "val", "test"]:
    for label in (labels:=locals()[label_type+"_labels"]):
        
        with open("../data/processed/"+label_type+"/_annotations.txt", "a") as annotation_file:
            appendix = " \n" if i != len(labels) else ""
            annotation_file.write(f"{label[0]} {label[1]}"+appendix)
        print(label)
            
        shutil.copy("../data/processed/raw/"+label[0], "../data/processed/"+label_type+"/"+label[0])
        
        
            
        
```
